```python
import openai
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

openai.api_key  = os.getenv('OPENAI_API_KEY')

def get_completion(messages, model="gpt-3.5-turbo"):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=1.25,
        max_tokens=100,
    )
    return response.choices[0].message["content"]

def get_completion_from_messages(messages, model="gpt-3.5-turbo", temperature=1.25, max_tokens=500):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens
    )
    return response.choices[0].message["content"]

def personal_trainer():
    prompt = """I want you to act as a professional fitness trainer. To establish the topic and name of the user, you will ask for the user name and the type of fitness program they are interested in.

You will create a 'physical and psychological questionnaire' to understand me before recommending a fitness program.

After I answer the questionnaire, you can ask 'clarifying questions' or 'additional questions' to gain a deeper understanding of me. You can ask the clarifying questions after all questionnaire questions have been answered.

The questionnaire is no more than 6 questions. You will ask the questionnaire questions one question at a time before moving onto the next question.

For the clarifying questions, you can ask no more than 3 questions.  You can ask all the clarifying questions at once.

In the end, you will create a highly detailed fitness plan for me. The plan will include detailed directions with a schedule so that I may build a healthy habit of exercising daily."""
    print("Hi, I am Katana, your AI personal trainer. What is your name?")
    user_input = input("> ")
    messages = [
        {"role": "system", "content": prompt},
    ]

    messages.append({"role": "user", "content": user_input})
    context = get_completion(messages)

    while True:
        user_input = input("> ")
        messages.append({"role": "user", "content": user_input})
        response = get_completion_from_messages(messages)
        print(response)
        messages.append({"role": "system", "content": response})
        context += "\n" + user_input + "\n" + response

personal_trainer()
```
